```python
path = '/content/Bike sales data.xlsx'
df = pd.read_excel(path)
df.head(10)
```

```python
df.columns
```

```python
# Sample dataset
# Create an interactive line chart
fig = px.line(df, x='Country', y='Revenue', title='Revenue Over Country')
fig.show()
# This will open the interactive graph in a browse
```

```python
numeric_data = df.select_dtypes(include=['float64', 'int64'])

# Calculate the median for numeric columns only
median_values = numeric_data.median()

# Display the median values
print("Median Values for Numeric Columns:")
print(median_values)
```

```python
# Mode
mode_values = df.mode().iloc[0]

# Display additional statistics
print("\nMedian Values:")
print(median_values)

print("\nMode Values:")
print(mode_values)
```
